# Analyse

Considérant que les 3 modèles montrent une performance pratiquement identique, on choisit de garder CatBoost comme modèle final en raison de sa simplicité et le fait qu'il gère nativement les données catégorique. Étant donné tout autre paramètre égaux, on opte donc pour le modèle le plus simple.

En se basant sur la recherche en grille faite dans 'agregation_xgb-svc-knn-cat.ipynb', on note les hyperparamètres pour obtenir ce classifieur, soit:

- depth : 6
- iterations : 1000
- l2_leaf_reg : 3
- learning_rate : 0.01
- one_hot_max_size : 150

## Analyse d'importance des features
On peut maintenant évaluer, à l'aide d'un classifieur CatBoost entraîné avec ces hyperparamètres, les features du dataset simplifié qui ont le plus d'importance

```python
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier

from load_dataset import load_data

import utils

X, y, cat_features_catb = load_data().get_data_X_y(data='simplify')
X_train_simplified, X_test_simplified, y_train_simplified, y_test_simplified = train_test_split(X, y, test_size=0.3, random_state=42)

catboost = CatBoostClassifier(cat_features=cat_features_catb, depth=6, iterations=1000, l2_leaf_reg=3, learning_rate=0.01, one_hot_max_size=150, verbose=0, random_state=42)
catboost.fit(X_train_simplified, y_train_simplified)

utils.features_importance(catboost, X_train_simplified.columns, plot=True)
```
