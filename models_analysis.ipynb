# Compilation
Les modèles sont compilés dans Agregation-DT-RF-CNB-LR.ipynb et agregation_xgb-svc-knn-cat.ipynb. On y compare les différents modèles et l'évaluation se fait principalement selon leur précision (accuracy) et leur performance (f1-score), qui sont les métriques choisies pour comparer les modèles.

Pour la sélection finale, nous nous concentrons principalement sur les résultats avec les données simplifiées puisque celles-ci éliminent certaines features qui ne sont pas pertinentes pour notre objectif, qui est de déterminer l'influence que les circonstances d'un étudiant au moment de son inscription auront sur ses chances de succès scolaire. Ainsi, certaines features comme les notes de l'étudiant au cours de son parcours furent éliminées du dataser simplifié.
